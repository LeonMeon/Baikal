```python
power = 2
bias_mask = False

class ResNet_Block_IN(torch.nn.Module):
    def __init__(self,input_size):
        super().__init__()
        self.module = torch.nn.Sequential(    
            torch.nn.Conv1d(input_size, input_size*power,  kernel_size=3, stride=1,padding=1,bias = bias_mask),
            torch.nn.BatchNorm1d(input_size*power),
            torch.nn.PReLU())
    def forward(self, inputs):
          return self.module(inputs) 
class ResNet_Block_OUT(torch.nn.Module):
    def __init__(self,input_size):
        super().__init__()
        self.module = torch.nn.Sequential(    
            torch.nn.Conv1d(input_size*power, input_size,  kernel_size=3, stride=1,padding=1,bias = bias_mask),
            torch.nn.BatchNorm1d(input_size),
            torch.nn.PReLU())
    def forward(self, inputs):
          return  self.module(inputs) 
        
class Inter_block(torch.nn.Module):
    def __init__(self,input_size):
        super().__init__()
        self.module = torch.nn.Sequential(    
            torch.nn.Conv1d(input_size*power, input_size*power,  kernel_size=3, stride=1,padding=1,bias = bias_mask),
            torch.nn.BatchNorm1d(input_size*power),
            torch.nn.PReLU())
        self.conv = torch.nn.Sequential( 
            torch.nn.Conv1d(input_size*power, input_size*power, kernel_size =1,bias =  bias_mask),
            torch.nn.PReLU())
    def forward(self, inputs):
          return  (self.module(inputs) +self.conv(inputs))
class Res_conv_block(torch.nn.Module):
    def __init__(self,input_size):
        super().__init__()
        self.conv = torch.nn.Sequential( 
            torch.nn.Conv1d(input_size*power, input_size*power, kernel_size =1,bias =  bias_mask),
            torch.nn.PReLU())        
    def forward(self, inputs):
          return  self.conv(inputs) 
class ResNet_Block_power_v4(torch.nn.Module):
    def __init__(self,input_size):
        super().__init__()
        self.start = ResNet_Block_IN(input_size)
        self.end = ResNet_Block_OUT(input_size)
        self.inter_module_1, self.inter_module_2 = Inter_block(input_size), Inter_block(input_size)
        self.inter_module_3, self.inter_module_4 = Inter_block(input_size), Inter_block(input_size)
        self.inter_module_5, self.inter_module_6 = Inter_block(input_size), Inter_block(input_size)
        self.res_1_2 , self.res_1_3, self.res_2_3 = Res_conv_block(input_size), Res_conv_block(input_size), Res_conv_block(input_size)
        self.res_1_4 , self.res_2_4, self.res_3_4 = Res_conv_block(input_size), Res_conv_block(input_size), Res_conv_block(input_size)
        self.res_1_5 , self.res_2_5, self.res_3_5 = Res_conv_block(input_size), Res_conv_block(input_size), Res_conv_block(input_size)
        self.res_4_5 , self.res_1_6, self.res_2_6 = Res_conv_block(input_size), Res_conv_block(input_size), Res_conv_block(input_size)
        self.res_3_6 , self.res_4_6, self.res_5_6 = Res_conv_block(input_size), Res_conv_block(input_size), Res_conv_block(input_size)
        
    def forward(self, inputs):
        x_1 = self.start(inputs)
        x_inter_1 = self.inter_module_1(x_1)
        x_inter_2 = self.inter_module_2(x_inter_1)+ self.res_1_2(x_1)
        x_inter_3 = self.inter_module_3(x_inter_2) + self.res_2_3(x_inter_1) + self.res_1_3(x_1) 
        x_inter_4 = self.inter_module_4(x_inter_3) + self.res_3_4(x_inter_2) + self.res_2_4(x_inter_1) + self.res_1_4(x_1) 
        x_inter_5 = self.inter_module_5(x_inter_4) + self.res_4_5(x_inter_3) + self.res_3_5(x_inter_2) + self.res_2_5(x_inter_1) + self.res_1_5(x_1)
        x_inter_6 = self.inter_module_6(x_inter_5) + self.res_5_6(x_inter_4) + self.res_4_6(x_inter_3) + self.res_3_6(x_inter_2) + self.res_2_6(x_inter_1) + self.res_1_6(x_1)
        x_out = self.end(x_inter_6)
        return  x_out      

a = torch.ones((512,5 ,32))
net_MC_data_v4 = torch.nn.Sequential(    
    torch.nn.Conv1d(5, 5,  kernel_size= 3 , stride = 1 ,padding= 1 ,bias = bias_mask),
    torch.nn.BatchNorm1d(5),
    torch.nn.PReLU(),
    ResNet_Block_power_v4(5), #32
    Transition_Block(5), #16
    ResNet_Block_power_v4(10), #16
    Transition_Block(10), #8
    ResNet_Block_power_v4(20), 
    Transition_Block(20), #4
    ResNet_Block_power_v4(40), # 4
    torch.nn.Flatten(),
    torch.nn.Linear(160,160),
    torch.nn.BatchNorm1d(160),
    torch.nn.PReLU(),
    torch.nn.Linear(160,160),
    torch.nn.BatchNorm1d(160),
    torch.nn.PReLU(),
    torch.nn.Linear(160 ,2)
) 


print(net_MC_data_v4(a).shape )
sum(p.numel() for p in net_MC_data_v4.parameters()   )


```
